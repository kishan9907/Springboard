```python
m = folium.Map(location=[45.46440063879739, 9.194880739442098],tiles="Stamen Toner" #"cartodbpositron"
               , zoom_start=12)

tooltip=GeoJsonTooltip(
    fields=["cellId", "SMS"],
    aliases=["cellId", "SMS"],
    localize=True,
    sticky=False,
    labels=True,
    style="""
        background-color: #F0EFEF;
        border: 2px solid black;
        border-radius: 3px;
        box-shadow: 3px;
    """,
    max_width=800,
)

popup=GeoJsonPopup(
    fields=["cellId", "SMS"], 
    aliases=["cellId", "SMS"], 
    localize=True, 
    labels=True, 
    style="background-color: yellow;")

g = folium.GeoJson(
    mergedf[['cellId','SMS','geometry']],
    name = 'SMSs',
    style_function=lambda x: {
        'fillColor': colormap(sdf_dict[x['properties']['cellId']]),
        'color': 'black',
        'weight': 0.4,
#         'dashArray': '5, 5',
        'fillOpacity': 0.8,
        'line_opacity':0.2,
    },
    tooltip=tooltip,
    popup=popup
).add_to(m)

folium.LayerControl().add_to(m)

m.save('AllGrids.html')
```

```python
timedf = cdf2[['SMS']].reset_index()
```

```python
timedf
```

```python
timedf["timeIntervalts"] = (timedf["timeIntervalts"].astype(int)// 10**9).astype('U10')
timedf.head() 
```

```python
%%time
data_to_color = "SMS"
cells = list(timedf["cellId"].unique())
dates = list(timedf["timeIntervalts"].unique())
colors = get_hex_colors.get_hex_colors(timedf, data_to_color, cmap = matplotlib.cm.Reds)#, log = True) # color dict that get's created from our get_hex_colors based on the values you want to plot
timedf["COLORS"] = timedf[data_to_color].map(colors) # create a colum in the df so that we can iterate and create the styledict
```

```python
styledict = {}
for cell in cells:
    #print(row)
    styledict[str(cell)] = {row['timeIntervalts']: {'color': row['COLORS'],
                                   'opacity': 0.8,'weight': 0.4,'fillOpacity': 0.8,'line_opacity':0.2,} 
                            for index, row in timedf[timedf['cellId']==cell].iterrows()}
```

```python
len(styledict)
```

```python
# creates and renders the Folium map
m = folium.Map(location=[45.46440063879739, 9.194880739442098],tiles="Stamen Toner" #"cartodbpositron"
               , zoom_start=12)

g = TimeSliderChoropleth(
    gdf.set_index("cellId").to_json(), # get's the coordinates for each id 
    styledict = styledict, # styledict contains for each id the timestamp and the color to plot.
    name = 'TimeSlider',
    overlay = False
)

m.add_child(g)
m.save('TimeSliderChoropleth1.html')

m
```
